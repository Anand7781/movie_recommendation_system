```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id' , 'title' , 'overview' , 'genres' , 'keywords' , 'cast' , 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast

def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l


movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast

def get_director(obj):
    try:
       
        if isinstance(obj, str):
            obj = ast.literal_eval(obj)
    except Exception:
        return []  
    
    for i in obj:
        if isinstance(i, dict) and i.get('job') == 'Director':
            return [i['name']]
    
    return []
    
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","")for i in x])
movies.head()
```

```python
def join_tags(row):
    return ' '.join(
        str(row['overview']).split() + 
        row['genres'] + 
        row['keywords'] + 
        row['cast'] + 
        row['crew']
    )

movies['tags'] = movies.apply(join_tags, axis=1)
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# stemin technique = [loved , loving , love] is convetrd into [love , love , love]
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i)) 
    return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0]))reverse = True , key=lambda x:x[1][1:6])
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]  # skip 0th (same movie)
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
